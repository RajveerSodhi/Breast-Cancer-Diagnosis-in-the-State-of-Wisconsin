```python
print("Columns in the dataset")
df.columns
```

```python
fig=sns.countplot(x='Diagnosis', data=df)
plt.title('Diagnosis Count')
fig.bar_label(fig.containers[0], label_type='center', size=13)

```

**This shows that most of the diagnosis are benign**

```python
features = df[['Diagnosis', 'Mean Perimeter', 'Mean Radius', 'Mean Texture']]
fig=sns.pairplot(data=features, hue='Diagnosis')

plt.title("Comparison of Mean Texture, Perimeter and Radius for different types of Diagnosis",x=-0.35,y=-0.5)
plt.show()
```

**It can be inferred that Malignant tumors have higher values for the radius, texture and perimeter features compared to 
benign tumors.**

```python
fig=sns.jointplot(x='Mean Radius', y='Diagnosis', data=df, height=8)
fig.ax_joint.set_yticklabels(['Malignant', 'Benign'])
plt.suptitle('Relationship between mean radius and diagnosis type', y=1.05)
plt.show()
```

**On analyzing the joint plot, it can be identified that most Malignant diagnosis  have a higher radius, while most Benign diagnosis have lower radius.**

```python
sns.violinplot(x='Diagnosis', y='Mean Area', data=df)
plt.title('Distribution of Area of tumour by diagnosis')
plt.xlabel('Type of Diagnosis')
plt.ylabel('Mean Area')
plt.show()
```

**It can be understood that the area of the tumour is wider for malignant cases and narrow for the benign cases**

```python
fig=sns.pairplot(df[["Mean Concavity", "Mean Smoothness", "Mean Texture","Diagnosis"]],hue="Diagnosis")
plt.suptitle("Comparison of Mean Texture, Perimeter and Radius for different types of Diagnosis",y=1.05)

plt.show()
```

**On analyzing the pairplots, we can notice that a weak negative association between the two features can be seen in the scatter plot between the concavity and smoothness values. As the concavity value decreases, the smoothness value increases**

**According to the diagonal graph, it is seen that the smoothness features has a normal distribution, the concavity and texture features have distributions that are right-skewed.**

## **Analysis**

As our EDA is now complete, we can analyze our data through the EDA to try answering our research questions.

### 1. Based on the features of several key features (for example radius and texture), can we accurately identify whether the tumour being benign or malignant?




To identify whether the tumour is benign or malignant through key features, we can use a variety of visualization techniques to understand if there are any patterns between the type of diagnosis and features.

```python
sns.set_theme(style="dark")
sns.jointplot(x='Mean Radius', y='Mean Texture', data=df, hue='Diagnosis', kind='kde')
```

When we analyze the given jointplot using a KDE counter plot, we can see that for Malignant tumours, the contour plot is more clustered with a higher peak. Furthermore, the counter plot is spread out with a lower peak for Benign tumours. 

On further analysis of the marginal histograms, we can see that the radius of malignant tumours is skewed more towards the right (higher values) compared to benign tumours. We can notice the same observation for the mean texture between malignant and benign tumours.

```python
sns.scatterplot(data=df, x='Mean Area', y='Mean Concavity', hue='Diagnosis', s=50)
sns.histplot(data=df, x='Mean Area', y='Mean Concavity', bins=50, pthresh=.1, cmap="mako",color='g')
sns.kdeplot(data=df, x='Mean Area', y='Mean Concavity', levels=5, color="r", linewidths=1)
```

We can notice that Benign tumors are clustered towards the lower end of the concavity and area features, whereas malignant tumors are distributed across a broader range of values. This shows that malignant tumours and benign tumours are closely interconnected in terms of the area and the concavity feature.

```python
sns.pairplot(data=df, hue='Diagnosis', vars=['Mean Area', 'Mean Smoothness', 'Mean Compactness'], corner=True)
plt.show()
```

Through this, we can see that the distribution of the area for malignant tumors appears to be skewed towards higher values compared to benign tumors. Furthermore, the distribution of smoothness and symmetry for benign tumors appears to be slightly skewed towards lower values compared to malignant tumors.
