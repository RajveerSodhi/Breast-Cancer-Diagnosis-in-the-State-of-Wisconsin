## Exploratory Data Analysis (EDA)


```python
print("Columns in the dataset")
df.columns
```

```python
print("Basic information regarding the dataset")
df.info()
```

```python
print("Description of the dataset")
df.describe()
```

```python
fig=sns.countplot(x='Diagnosis', data=df)
plt.title('Diagnosis Count')
fig.bar_label(fig.containers[0], label_type='center', size=13)

```

**This shows that most of the diagnosis are benign**

```python
features = df[['Diagnosis', 'Mean Perimeter', 'Mean Radius', 'Mean Texture']]
fig=sns.pairplot(data=features, hue='Diagnosis')

plt.title("Comparison of Mean Texture, Perimeter and Radius for different types of Diagnosis",x=-0.35,y=-0.5)
plt.show()
```

**It can be inferred that Malignant tumors have higher values for the radius, texture and perimeter features compared to 
benign tumors.**

```python
fig=sns.jointplot(x='Mean Radius', y='Diagnosis', data=df, height=8)
fig.ax_joint.set_yticklabels(['Malignant', 'Benign'])
plt.suptitle('Relationship between mean radius and diagnosis type', y=1.05)
plt.show()
```

**On analyzing the joint plot, it can be identified that most Malignant diagnosis  have a higher radius, while most Benign diagnosis have lower radius.**

```python
sns.violinplot(x='Diagnosis', y='Mean Area', data=df)
plt.title('Distribution of Area of tumour by diagnosis')
plt.xlabel('Type of Diagnosis')
plt.ylabel('Mean Area')
plt.show()
```

**It can be understood that the area of the tumour is wider for malignant cases and narrow for the benign cases**

```python
fig=sns.pairplot(df[["Mean Concavity", "Mean Smoothness", "Mean Texture","Diagnosis"]],hue="Diagnosis")
plt.suptitle("Comparison of Mean Texture, Perimeter and Radius for different types of Diagnosis",y=1.05)

plt.show()
```

**On analyzing the pairplots, we can notice that a weak negative association between the two features can be seen in the scatter plot between the concavity and smoothness values. As the concavity value decreases, the smoothness value increases**

**According to the diagonal graph, it is seen that the smoothness features has a normal distribution, the concavity and texture features have distributions that are right-skewed.**
