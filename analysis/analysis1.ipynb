```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("../data/raw/data.csv")
df
```

On analysing the different parameters found in the dataset, it can be said that the "id" and "Unnamed:32" columns are not 
required for our analysis.


```python
df_new = df.copy().drop(['id','Unnamed: 32'], axis=1)
df_new
```

```python
df_new.describe(exclude=np.number)
```

```python
df_new.describe().apply(lambda s: s.apply(lambda x: format(x, 'f')))
```

```python
sns.countplot(x='diagnosis', data=df_new)
plt.title('Diagnosis Count')
```

**This shows that most of the diagnosis are balignant**

```python
sns.histplot(data=df_new, x='radius_mean', hue='diagnosis', kde=True)
```

```python
sns.histplot(data=df_new, x='texture_mean', hue='diagnosis', kde=True)
```

```python
sns.histplot(data=df_new, x='perimeter_mean', hue='diagnosis', kde=True)
```

**It can be inferred that Malignant tumors have higher values for the radius, texture and perimeter features compared to 
benign tumors.**

```python
sns.jointplot(x='radius_mean', y='diagnosis', data=df_new, height=8)
plt.show()
```

**On analyzing the joint plot, it can be identified that most Malignant diagnosis  have a higher radius, while most Benign diagnosis have lower radius.**

```python
sns.violinplot(x='diagnosis', y='area_mean', data=df_new)
plt.title('Distribution of Area of tumour by diagnosis')
plt.show()
```

**It can be understood that the area of the tumour is wider for malignant cases and narrow for the benign cases**

```python
sns.pairplot(df[["concavity_mean", "smoothness_mean", "texture_mean","diagnosis"]],hue="diagnosis")
plt.show()
```

**On analyzing the pairplots, we can notice that a weak negative association between the two features can be seen in the scatter plot between the concavity and smoothness values. As the concavity value decreases, the smoothness value increases**

**According to the diagonal graph, it is seen that the smoothness features has a normal distribution, the concavity and texture features have distributions that are right-skewed.**
